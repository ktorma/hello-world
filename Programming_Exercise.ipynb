# Task 1

## 5_Net asset valuation of the portfolio or the share class in portfolio currency
### According to the TPT definition doc, we have 28_Market exposure amount in portfolio currency (B) and 30_Market Exposure in weight = Exposure valuation in portfolio currency / total net asset value of the fund, in % . I set then Field 5 = SUM (Field 28) since SUM(Field 30)=1 (I would have set Field 5= Field 28/ Field 30, but this would not give the same number for all the holdngs in the portfolio). Portfolio currency according to 4_Portfolio currency ( B ) is in USD.

```python
Table_TPT_change['5_Portfolio-TotalNetAssets'] = Table_TPT_change['28_Position-Valuation-MarketExposurePC'].sum()
display(Table_TPT_change)
```

## 8b_Portfolio-ShareClass-TotalNumberOfShares
### According to the TPT definition doc, (Field 8 * field 8b = 5)

```python
Table_TPT_change['8b_Portfolio-ShareClass-TotalNumberOfShares'] = Table_TPT_change['5_Portfolio-TotalNetAssets']/Table_TPT_change['8_Portfolio-ShareClass-SharePrice']
display(Table_TPT_change)
```

## Calculate the total portfolio value in CHF.

### I find the FX conversion USD/CHF by relating 24_Position-Valuation-MarketValuePC (always in USD) and 22_Position-Valuation-MarketValueQC for any of the cash items whose quotation currency is CHF, for example 14_Position-InstrumentCode-Code = XT4944922406, for which 21_Position-Valuation-QuotationCurrency = CHF

```python
FX_CHF = (Table_TPT_change['22_Position-Valuation-MarketValueQC'][(Table_TPT_change['12_Position-InstrumentCIC'] == 'XT71')
                                                                        & (Table_TPT_change['21_Position-Valuation-QuotationCurrency'] == 'CHF')][0]/
                Table_TPT_change['24_Position-Valuation-MarketValuePC'][(Table_TPT_change['12_Position-InstrumentCIC'] == 'XT71')
                                                                        & (Table_TPT_change['21_Position-Valuation-QuotationCurrency'] == 'CHF')][0])

total_PFV_CHF = FX_CHF * Table_TPT_change['5_Portfolio-TotalNetAssets'][0]
    
print total_PFV_CHF
```

# Task 2

## Total value of the cash items in USD. Cash items are all those identified by CIC code XT71, thus I sum 24_Position-Valuation-MarketValuePC for all of them 

```python
ALL_CASH_USD = Table_TPT_change['24_Position-Valuation-MarketValuePC'][Table_TPT_change['12_Position-InstrumentCIC'] == 'XT71'].sum()
print 'Total Value of Cash in USD = '+ str( ALL_CASH_USD)
```

## Currency distribution in the portfolio in percentage. From 21_Position-Valuation-QuotationCurrency we can observe two currencies, USD and CHF, thus we need the proportion in USD of CHF items and the proportion in USD of CHF items

```python
CHF_PERCENTAGE = Table_TPT_change['24_Position-Valuation-MarketValuePC'][Table_TPT_change['21_Position-Valuation-QuotationCurrency'] == 'CHF'].sum()/\
                            Table_TPT_change['24_Position-Valuation-MarketValuePC'].sum()
    
print 'CHF percentage = '+str( "{:.2%}".format(CHF_PERCENTAGE) )
    
USD_PERCENTAGE = Table_TPT_change['24_Position-Valuation-MarketValuePC'][Table_TPT_change['21_Position-Valuation-QuotationCurrency'] == 'USD'].sum()/\
                            Table_TPT_change['24_Position-Valuation-MarketValuePC'].sum()
    
print 'USD percentage = '+str( "{:.2%}".format(USD_PERCENTAGE) )

print "{:.2%}".format(CHF_PERCENTAGE + USD_PERCENTAGE)
```

## 9_Portfolio-CashPercentage: Amount of cash of the fund / total net asset value of the fund, in %. ALL_CASH_USD has been calculated in the first point, we just need to divide by the total in Field 5

```python
TOTAL_CASH_PERCENTAGE = ALL_CASH_USD/Table_TPT_change['5_Portfolio-TotalNetAssets'][0]

print 'Total Cash Percentage = '+str( "{:.2%}".format(TOTAL_CASH_PERCENTAGE) )

Table_TPT_change['9_Portfolio-CashPercentage'] = "{:.2%}".format(TOTAL_CASH_PERCENTAGE)

display(Table_TPT_change)
```

# Task 3

## a) list of the position ids (14_Position-InstrumentCode-Code) which is in descending order by maturity (39_Position-IntRateInst-Redemption-MaturityDate)

```python
Table_TPT_MAT = Table_TPT_change[['14_Position-InstrumentCode-Code','39_Position-IntRateInst-Redemption-MaturityDate']][~Table_TPT_change['39_Position-IntRateInst-Redemption-MaturityDate'].isnull()]

Table_TPT_MAT = Table_TPT_MAT.sort_values(by='39_Position-IntRateInst-Redemption-MaturityDate', ascending=False)

display(Table_TPT_MAT)

LIST_a = list(Table_TPT_MAT['14_Position-InstrumentCode-Code'].astype(str))[:10]

print 'First 10 elements: '+str( LIST_a )
```

## b) ...

```python
Table_TPT_MAT2 = Table_TPT_MAT.copy()

Table_TPT_MAT2['39_Position-IntRateInst-Redemption-MaturityDate'] = pd.to_datetime(Table_TPT_MAT2['39_Position-IntRateInst-Redemption-MaturityDate'])

Table_TPT_MAT2['Year'] = Table_TPT_MAT2['39_Position-IntRateInst-Redemption-MaturityDate'].dt.year
Table_TPT_MAT2['Month'] = Table_TPT_MAT2['39_Position-IntRateInst-Redemption-MaturityDate'].dt.month
Table_TPT_MAT2['Month_c'] = 13 - Table_TPT_MAT2['39_Position-IntRateInst-Redemption-MaturityDate'].dt.month
Table_TPT_MAT2['Month_c'][Table_TPT_MAT2['Month_c'] < 10] = '0' + Table_TPT_MAT2['Month_c'].astype(str)
Table_TPT_MAT2['Day'] = Table_TPT_MAT2['39_Position-IntRateInst-Redemption-MaturityDate'].dt.day
Table_TPT_MAT2['Date_c'] = (Table_TPT_MAT2['Year'].astype(str)+Table_TPT_MAT2['Month_c'].astype(str)+Table_TPT_MAT2['Day'].astype(str)).astype(int)

Table_TPT_MAT2 = Table_TPT_MAT2.sort_values(by='Date_c', ascending=True)

display(Table_TPT_MAT2)

LIST_b = list( Table_TPT_MAT2['14_Position-InstrumentCode-Code'].astype(str) )[-10:]

print 'Last 10 elements' + str(LIST_b)

print 'Final List = '+str(LIST_a + LIST_b)
```

# Task 4

```python
STRESS_UP = Table_TPT_change['97_Position-ContributionToSCR-MktIntUp']/Table_TPT_change['24_Position-Valuation-MarketValuePC']
STRESS_DOWN = Table_TPT_change['98_Position-ContributionToSCR-MktintDown']/Table_TPT_change['24_Position-Valuation-MarketValuePC']

STRESS_UP_mean = STRESS_UP.mean()
STRESS_UP_median = STRESS_UP.median()
STRESS_UP_sorted = Numpy.sort(STRESS_UP)
len_STRESS_UP = len(STRESS_UP_sorted)

STRESS_UP_05 = STRESS_UP_sorted[int(0.05*len_STRESS_UP)]
STRESS_UP_10 = STRESS_UP_sorted[int(0.1*len_STRESS_UP)]
STRESS_UP_25 = STRESS_UP_sorted[int(0.25*len_STRESS_UP)]
STRESS_UP_75 = STRESS_UP_sorted[int(0.75*len_STRESS_UP)]
STRESS_UP_90 = STRESS_UP_sorted[int(0.90*len_STRESS_UP)]
STRESS_UP_95 = STRESS_UP_sorted[int(0.95*len_STRESS_UP)]

STRESS_UP_STD = STRESS_UP.std()

print 'Mean UP= '+str(STRESS_UP_mean)
print 'Median UP = '+str(STRESS_UP_median)
print 'Percentiles 5, 10, 25, 75, 90, 95  UP = '+str(STRESS_UP_05)+', '+str(STRESS_UP_10)+', '+str(STRESS_UP_25)+', '+str(STRESS_UP_75)+', '+str(STRESS_UP_90)+', '+str(STRESS_UP_95)
print 'Standard Deviation UP = '+str(STRESS_UP_STD)

STRESS_DOWN_mean = STRESS_DOWN.mean()
STRESS_DOWN_median = STRESS_DOWN.median()
STRESS_DOWN_sorted = -Numpy.sort(-STRESS_DOWN)
len_STRESS_DOWN = len(STRESS_DOWN_sorted)

STRESS_DOWN_05 = STRESS_DOWN_sorted[int(0.05*len_STRESS_DOWN)]
STRESS_DOWN_10 = STRESS_DOWN_sorted[int(0.1*len_STRESS_DOWN)]
STRESS_DOWN_25 = STRESS_DOWN_sorted[int(0.25*len_STRESS_DOWN)]
STRESS_DOWN_75 = STRESS_DOWN_sorted[int(0.75*len_STRESS_DOWN)]
STRESS_DOWN_90 = STRESS_DOWN_sorted[int(0.90*len_STRESS_DOWN)]
STRESS_DOWN_95 = STRESS_DOWN_sorted[int(0.95*len_STRESS_DOWN)]

STRESS_DOWN_STD = STRESS_DOWN.std()

print 'Mean DOWN= '+str(STRESS_DOWN_mean)
print 'Median DOWN = '+str(STRESS_DOWN_median)
print 'Percentiles 5, 10, 25, 75, 90, 95  DOWN = '+str(STRESS_DOWN_05)+', '+str(STRESS_DOWN_10)+', '+str(STRESS_DOWN_25)+', '+str(STRESS_DOWN_75)+', '+str(STRESS_DOWN_90)+', '+str(STRESS_DOWN_95)
print 'Standard Deviation DOWN = '+str(STRESS_DOWN_STD)
```
